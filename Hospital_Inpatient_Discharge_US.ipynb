# Health Equity & Access Analysis:-

```python
urban_counties = [
    'New York', 'Kings', 'Queens', 'Bronx', 'Richmond',  # NYC
    'Nassau', 'Suffolk', 'Westchester', 'Erie', 'Monroe', 'Albany'
]

df['County Type'] = df['Hospital County'].apply(
    lambda x: 'Urban' if x in urban_counties else 'Rural'
)
```

```python
# Clean data first
df['Total Charges'] = df['Total Charges'].replace('[\$,]', '', regex=True).astype(float)
df['Length of Stay'] = pd.to_numeric(df['Length of Stay'], errors='coerce')

# Group by County Type
urban_rural_stats = df.groupby('County Type')[['Total Charges', 'Length of Stay']].mean()
print(urban_rural_stats)
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='County Type', y='Total Charges', data=df)
plt.title("Total Charges: Urban vs Rural")
plt.show()

plt.figure(figsize=(8,5))
sns.boxplot(x='County Type', y='Length of Stay', data=df)
plt.title("Length of Stay: Urban vs Rural")
plt.show()
```

```python
df['Total Costs'] = df['Total Costs'].replace('[\$,]', '', regex=True).astype(float)
df['Total Charges'] = df['Total Charges'].replace('[\$,]', '', regex=True).astype(float)


# cumulative totals
df['Cumulative Costs'] = df['Total Costs'].cumsum()
df['Cumulative Charges'] = df['Total Charges'].cumsum()


plt.figure(figsize=(12, 6))
plt.plot(df.index, df['Cumulative Costs'], label='Cumulative Costs', color='red')
plt.plot(df.index, df['Cumulative Charges'], label='Cumulative Charges', color='green')
plt.xlabel('Number of Discharges (Progression)')
plt.ylabel('Cumulative Amount (in $)')
plt.title('Worm Plot: Total Charges vs Total Costs to Hospital')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
